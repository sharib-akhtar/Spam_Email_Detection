```python
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\SHARIB\\OneDrive\\Desktop\\spamEmail\\spam.csv")
```

```python
df.groupby('Category').describe()
```

```python
df['spam']=df['Category'].apply(lambda x:1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.Message,df.spam,test_size=0.2)
```

```python
x_train.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v=CountVectorizer()
```

```python
x_train_vect=v.fit_transform(x_train.Message)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB ()
model.fit(x_train_vect,y_train)
```

```python
Email=["Hey Mohan can we get together to watch cricket game tomorrow?",
"Buy this phone at a 50% discount and get 1 month recharge free."]
```

```python
emailVect=v.transform(Email)
```

```python
model.predict(emailVect)
```

```python
x_testVect=v.transform(x_test.Message)
model.score(x_testVect,y_test)
```
